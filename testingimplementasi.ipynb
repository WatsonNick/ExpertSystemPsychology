```python
import pandas as pd

# Read the data from the CSV file
data = pd.read_csv("Copy of fakta-diagnosis - Sheet1.csv")
data.set_index('No', inplace=True) # Show summary statistics

data.head(10)
```

```python
# Ambil fitur "ciri-ciri" dan ubah ke list
pertanyaan = data["Ciri-ciri"].tolist()
```

```python
# Dictionary untuk menyimpan data
data_kecerdasan_dict = {}

# Iterasi melalui baris DataFrame
for index, row in data.iterrows():
    statement = row['Ciri-ciri']
    category = row['Ciri-Ciri Kecerdasan']
    score = row['CF']  # Menggunakan kolom 'CF' sebagai skor

    if statement not in data_kecerdasan_dict:
        data_kecerdasan_dict[statement] = {}

    data_kecerdasan_dict[statement][category] = score

# Mengonversi ke format tuple
data_kecerdasan= [(statement, categories) for statement, categories in data_kecerdasan_dict.items()]

# Menampilkan hasil
for item in data_kecerdasan:
    print(item)
```

```python
# Dictionary untuk menyimpan data
data_prodi_dict = {}

# Iterasi melalui baris DataFrame
for index, row in data.iterrows():
    statement = row['Ciri-ciri']
    category = row['Prodi']
    score = row['CF']  # Menggunakan kolom 'CF' sebagai skor

    if statement not in data_prodi_dict:
        data_prodi_dict[statement] = {}

    data_prodi_dict[statement][category] = score

# Mengonversi ke format tuple
data_prodi= [(statement, categories) for statement, categories in data_prodi_dict.items()]

# Menampilkan hasil
for item in data_prodi:
    print(item)
```
